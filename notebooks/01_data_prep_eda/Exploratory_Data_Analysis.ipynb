## Grouping

```python
df_groceries.head()
```

```python
df_groceries.groupby('Category')['Inventory'].sum()
```

```python
df_groceries.groupby('Category')['Inventory'].count()
```

```python
df_groceries.groupby(['Category', 'Sub-Category'])['Inventory'].count().reset_index()
```

```python
df_groceries.groupby(['Category', 'Sub-Category'])['Inventory'].agg(['sum', 'count']).reset_index()
```

```python
df_groceries.head()
```

```python
#Incorrecto para el item con el maximo precio
df_groceries.groupby('Category')[['Item', 'Price_Dollars']].max()
```

```python
df_groceries.head()
```

```python
#Correcto para el item con el maximo precio
(df_groceries[['Category', 'Item', 'Price_Dollars']]
            .sort_values('Price_Dollars', ascending= False)
            .groupby('Category')
            .head(1))
```
