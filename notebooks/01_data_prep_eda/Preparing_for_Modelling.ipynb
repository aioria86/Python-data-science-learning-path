## Join Types

```python
happy_scores = pd.concat([ca, mx, us, cr.rename(columns={'nombre del país':'country_name'})]).reset_index(drop=True)
happy_scores = happy_scores[happy_scores.year > 2019]
happy_scores.head()
```

```python
pop = pd.read_excel('../../data/raw/List of countries and inhabited territories by total population. United Nations.xlsx')

#pop.groupby('Subregion')['Country'].count()
pop = pop[pop['UN statistical'] == 'Northern America']
pop


```

```python
happy_scores['country_name'] = happy_scores['country_name'].str.strip().str.lower()
pop['Country or territory'] = pop['Country or territory'].str.strip().str.lower()
```

```python
(happy_scores.merge(pop, 
                    how='inner', 
                    left_on='country_name', 
                    right_on='Country or territory'))
```

```python
(happy_scores.merge(pop, 
                    how='outer', 
                    left_on='country_name', 
                    right_on='Country or territory'))
```

```python
(happy_scores.merge(pop, 
                    how='left', 
                    left_on='country_name', 
                    right_on='Country or territory'))
```

```python
(happy_scores.merge(pop, 
                    how='right', 
                    left_on='country_name', 
                    right_on='Country or territory'))
```
